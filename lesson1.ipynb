```python
# f(x,y) = dict(x,y) -> f(name, age) = {'name': name, 'age': age}
# f(x,y) = 1 -> x^2 + y^2 = 1
def name_age_to_dict(name, age):
    d = {'name': name, 'age': age}
    return d

name = "Alice"
age = 10
name_age_to_dict(name, age)
```

```python
# 去超市买10个啤酒，3个啤酒瓶可以换一个新的啤酒，问最多可以买多少啤酒
full_bottle = 10 # 初识状态和更新状态的酒数量
empty_bottle = 0 # 当前酒瓶子的数量
total_beer = 0  # 累加一共喝了几瓶酒

while True:
    full_bottle = full_bottle - 1
    empty_bottle = empty_bottle + 1
    total_beer = total_beer + 1
    if empty_bottle >= 3: # 如果有三个空瓶子，就换一个啤酒
        empty_bottle = empty_bottle - 3 # 三个空瓶子换一瓶啤酒
        full_bottle = full_bottle + 1 # 新的啤酒加进去

    if full_bottle == 0: # 如果没有啤酒了，并且空瓶子不够换酒了，就结束
        break
    print("当前有", full_bottle,"瓶啤酒",empty_bottle,"个空瓶子，一共喝了",total_beer,"瓶啤酒")

print("总共喝了", total_beer, "瓶啤酒")
```
